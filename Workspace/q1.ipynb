```python
F, F_leave_base, F_arrive_base = data_loader.get_F()
F, F_leave_base, F_arrive_base
```

```python
nonF_leave_base = list(set(F) - set(F_leave_base[0]))
nonF_arrive_base= list(set(F) - set(F_arrive_base[0]))
```

```python
FF = data_loader.get_FF()
FF
```

```python
import gurobipy as gp
from gurobipy import *
```

```python
m=gp.Model('m1')

z=m.addVars(F,vtype=gp.GRB.BINARY,name='z')
x_ikdh=m.addVars(F,C,vtype=gp.GRB.BINARY)
x_ikfo=m.addVars(F,C,vtype=gp.GRB.BINARY)
u_iksta=m.addVars(F,C,vtype=gp.GRB.BINARY)
u_jkfin=m.addVars(F,C,vtype=gp.GRB.BINARY)
y_ijk=m.addVars(F,F,C,vtype=gp.GRB.BINARY)
x_ikcap=m.addVars(F,C,vtype=gp.GRB.BINARY)

m.ModelSense=GRB.MINIMIZE
m.setObjectiveN(-z.sum(),index=0,weight=0.8)
# m.setObjectiveN(x_ikdh.sum(),index=1,weight=0.15)
# m.setObjectiveN(x_ikfo.sum(),index=2,weight=0.05)

M=10000
m.addConstrs(u_iksta[i,k]==0 for i in nonF_leave_base for k in C)
m.addConstr(gp.quicksum(u_iksta[i,k] for i in F_leave_base[0] for k in C )<=1 )
m.addConstrs(x_ikfo[i,k]==0 for i in F for k in C1 )
m.addConstrs(x_ikcap[i,k]==0 for i in F for k in C3 )
m.addConstrs(u_jkfin[j,k]==0 for j in nonF_arrive_base for k in C)
m.addConstr(gp.quicksum(u_jkfin[j,k] for j in F_arrive_base[0] for k in C )<=1 )

m.addConstrs(y_ijk[i,j,k]==0 for i,j in FF for k in C)
m.addConstrs(x_ikcap[i,k]+x_ikfo[i,k]+x_ikdh[i,k]<=1 for i in F for k in C)
m.addConstrs(1-x_ikcap.sum(i,'*')<=M*(1-z[i]) for i in F )
m.addConstrs(1-x_ikfo.sum(i,'*')<=M*(1-z[i]) for i in F )
m.addConstrs(gp.quicksum(x_ikcap[i,k]+x_ikfo[i,k]+x_ikdh[i,k] for k in C)<=M*(z[i]) for i in F)
m.addConstrs(1-(y_ijk.sum(i,'*',k)+u_iksta[i,k]+u_jkfin[i,k])<=M*(1-(x_ikcap[i,k]+x_ikfo[i,k]+x_ikdh[i,k])) for i in F for k in C)
m.addConstrs(1-(y_ijk.sum('*',j,k)+u_iksta[j,k]+u_jkfin[j,k])<=M*(1-(x_ikcap[j,k]+x_ikfo[j,k]+x_ikdh[j,k])) for j in F for k in C)
m.addConstrs(M*z[j]>=gp.quicksum(x_ikcap[j,k]+x_ikfo[j,k]+x_ikdh[j,k] for k in C) for j in F)
m.update()
print('ok')
m.optimize()
```
